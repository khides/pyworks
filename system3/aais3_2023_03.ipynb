## ベータ分布：二項分布の共役事前分布

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst = [(1.0,1.0),(2.0,2.0),(10.0,10.0),(10.0,2.0),(10.0,1.0),(0.5,0.5)]
fig = plt.figure(figsize=(15,8))

th = np.linspace(0.0,1.0,100)
for i in range(6):
  ax = fig.add_subplot(2,3,i+1)
  a,b = lst[i]
  p = beta.pdf(th,a,b)
  ax.plot(th,p,'r-',lw=2)
  ax.set_title('alpha={:.1f}, beta={:.1f}'.format(a,b))
plt.show()
```

## コインを10回投げて7回表が出た場合

### 尤度 $p(D\vert \theta)$



```python
def likelihood_coin(th):
  return th**7 * (1-th)**3

lkhd = likelihood_coin(th)
plt.plot(th,lkhd,'r-')
```

### 事前分布をベータ分布で与えた場合の事後確率

```python
fig = plt.figure(figsize=(15,8))

for i in range(6):
  ax = fig.add_subplot(2,3,i+1)
  a,b = lst[i]
  p = beta.pdf(th,a,b) * lkhd
  ax.plot(th,p,'r-',lw=2)
  ax.set_title('alpha={:.1f}, beta={:.1f}'.format(a,b))
plt.show()
```
